```python
#first with import pandas and matplotlib scripting interface plt
import pandas as pd
import matplotlib.pyplot as plt
wine_counts = pd.read_csv("top-five-wine-score-counts.csv",index_col=0)
wine_counts.head(10)
```

```python
wine_counts.plot.bar(stacked=True)
```

```python
wine_counts.plot.area()
#area plots are used to show magutide and porportion of mutliple variables
#it is useful to compare multiple variables
#area plots are used on data with a cumultative nature
#this can be population or distrubutions
```

```python
#then we plot the area plot using scripting method
wine_counts.plot(kind='area',
            alpha=0.55,
             stacked=True,
             figsize=(20, 10))  # pass a tuple (x, y) size
plt.title('Wine counts based on points from 80-100')
plt.ylabel('Number of wine')
plt.xlabel('Points')
```

```python
#then we plot the area plot
wine_counts.plot(kind='area',
            alpha=0.45,
             stacked=False,
             figsize=(20, 10))  # pass a tuple (x, y) size

plt.title('Wine counts based on points from 80-100')
plt.ylabel('Number of wine')
plt.xlabel('Points')
```

```python
help(wine_counts)
```

```python
#plot using object orient
ax = wine_counts.plot(kind='area', alpha=0.35, figsize=(20, 10))

ax.set_title('Wine points scores')
ax.set_ylabel('Number of wine')
ax.set_xlabel('points')
print(wine_counts.info)
```

```python
#link of example
#https://seaborn.pydata.org/tutorial/objects_interface.html
#https://seaborn.pydata.org/generated/seaborn.objects.Area.html
import seaborn.objects as so
p = so.Plot(wine_counts,'points','Bordeaux-style Red Blend')
p.add(so.Area())
print(wine_counts.index)
print(wine_counts.values)
```
